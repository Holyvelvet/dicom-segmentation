```python
# You need to unzip files
import os
import zipfile


data_path = r'../data/01_raw'


def unpack_zip_files(data_path):
    
    zip_files = find_zips(data_path)
    
    if not zip_files:
        return
    
    for filename in zip_files:
        file_path = os.path.join(data_path, filename)
        folder_path = os.path.join(data_path, get_filename_without_extension(filename))
        
        os.makedirs(folder_path)
        
        print('Unpacking file {}'.format(filename))
        with zipfile.ZipFile(file_path, 'r') as zip_ref:
            zip_ref.extractall(folder_path)
    
        unpack_zip_files(folder_path)
    
        os.remove(file_path)

        
def find_zips(data_path):
    return [filename for filename in os.listdir(data_path) if filename.endswith('.zip')]

def get_filename_without_extension(filename):
    fn = os.path.splitext(filename)[0]
    
    if fn != filename:
        fn = get_filename_without_extension(fn)
        
    return fn
```
